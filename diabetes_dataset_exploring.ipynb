```python
import pandas as pd
```

```python
df = pd.read_csv('data/datasets/diabetic_data.csv')
```

1. Weight column is mostly "?"
2. Missing vals in max_glu_serum and A1Cresult columns - note: these 2 nulls can be a separate category - test not taken

```python
# Dropping id columns and weight, as it's mostly "?"
df.drop(columns=['encounter_id', 'patient_nbr', 'weight'], index=1, inplace=True)
```

```python
df
```

```python
df['readmitted'].value_counts()
```

```python
df['readmitted'] = df['readmitted'].map({'NO': 0, '>30': 1, '<30': 1})
```

```python
df['max_glu_serum'].value_counts()
```

```python
df['max_glu_serum'] = df['max_glu_serum'].replace("Norm", 0)
df['max_glu_serum'] = df['max_glu_serum'].replace(">200", 1)
df['max_glu_serum'] = df['max_glu_serum'].replace(">300", 2)
df['max_glu_serum'] = df['max_glu_serum'].fillna(-99)
df['max_glu_serum'].value_counts()
```

```python
df['A1Cresult'].value_counts()
```

```python
df['A1Cresult'] = df['A1Cresult'].replace("Norm", 0)
df['A1Cresult'] = df['A1Cresult'].replace(">7", 1)
df['A1Cresult'] = df['A1Cresult'].replace(">8", 2)
df['A1Cresult'] = df['A1Cresult'].fillna(-99)
df['A1Cresult'].value_counts()
```

```python
df
```

```python
df['age'].unique()
```

```python
df['age'] = df['age'].map({
    '[0-10)': 0,
    '[10-20)' : 1,
    '[20-30)' : 2,
    '[30-40)' : 3,
    '[40-50)' : 4,
    '[50-60)' : 5,
    '[60-70)' : 6,
    '[70-80)' : 7,
    '[80-90)' : 8,
    '[90-100)' : 9,
})

df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])
        
df
```

```python
df.to_csv('data/datasets/diabetic_data_transformed.csv', index=False)
```
